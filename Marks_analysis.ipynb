```python
# Step 1: Libraries import karna
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 2: CSV file read karna
data = pd.read_csv("marks.csv")
```

```python
# Step 3: Marks column nikalna
marks = data["Marks"]
```

```python
# Step 4: Mean aur Standard Deviation
mean_marks = marks.mean()
std_marks = marks.std()
```

```python
print("Mean Marks:", mean_marks)
print("Standard Deviation:", std_marks)
```

```python
# Step 5: Histogram banana
plt.hist(marks, bins=5)
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.title("Student Assignment Marks Distribution")

plt.show()
```
